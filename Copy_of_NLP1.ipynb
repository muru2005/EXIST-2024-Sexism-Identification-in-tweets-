```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
import numpy as np
vectorizer = TfidfVectorizer()

# Convert target variables to numerical values


# Assuming df_train and df_test are your two data frames
# Step 2: Split into features and labels
X_train = vectorizer.fit_transform(eng_tweets['tweet_wo_punc'])
 # Features for training
y_train = eng_tweets['classification']  # Target variable for training

X_test = vectorizer.transform(engtweet['tweet_wo_punc'])  # Features for testing
y_test = engtweet['classification']  # Target variable for testing

# Step 3: Further split into training and validation
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)



# Initialize models
svm_clf = SVC(probability=True)
rf_clf = RandomForestClassifier()
lr_clf = LogisticRegression()

# Train models
svm_clf.fit(X_train, y_train)
rf_clf.fit(X_train, y_train)
lr_clf.fit(X_train, y_train)

# Get predictions probabilities for each model
svm_proba = svm_clf.predict_proba(X_test)
rf_proba = rf_clf.predict_proba(X_test)
lr_proba = lr_clf.predict_proba(X_test)

# Soft voting ensemble: averaging probabilities
ensemble_proba = (svm_proba + rf_proba + lr_proba) / 3

# Convert probabilities to predictions
ensemble_predictions = np.argmax(ensemble_proba, axis=1)
ensemble_accuracy = accuracy_score(y_test, ensemble_predictions)

print("Ensemble Accuracy:", ensemble_accuracy)
y_pred = voting_clf.predict(X_test)

# Generate classification report
report = classification_report(y_test, y_pred)

# Print the classification report
print(report)

```
